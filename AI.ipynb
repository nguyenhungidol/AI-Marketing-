## Phần 1

```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr, spearmanr
import matplotlib.pyplot as plt

INPUT_CSV = "data_clean.csv"     

def read_and_clean(path):
    df = pd.read_csv(path, dtype=str)

    if "Tên sản phẩm" in df.columns:
        df["Tên sản phẩm"] = df["Tên sản phẩm"].str.strip()

    numeric_cols = [
        "Giá bán hiện tại","Giá gốc","Giảm giá",
        "Số lượng đã bán",
        "Số lượt xem sản phẩm",
        "Số lượng hàng tồn kho",
        "Điểm đánh giá",
        "Số lượng đánh giá",
        "Giá trị định lượng",
        "Định lượng (g)",
        "Định lượng (ml)"
    ]

    for c in numeric_cols:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")

    return df


def compute_metrics(df):
    df["Tổng doanh thu"] = df["Giá bán hiện tại"] * df["Số lượng đã bán"]

    df["Biên lợi nhuận"] = (df["Giá bán hiện tại"] - df["Giá gốc"]) / df["Giá gốc"]

    df["Lãi(+) / Lỗ(-) trên 1 sp"] = df["Giá bán hiện tại"] - df["Giá gốc"]

    df["Tổng lãi/lỗ"] = df["Lãi(+) / Lỗ(-) trên 1 sp"] * df["Số lượng đã bán"]

    # 🔥 MỚI – tỷ lệ chuyển đổi
    df["Tỷ lệ chuyển đổi"] = (
        df["Số lượng đã bán"] / df["Số lượt xem sản phẩm"]
    ).replace([np.inf, -np.inf], np.nan)

    # 🔥 MỚI – áp lực tồn kho
    df["Áp lực tồn kho"] = (
        df["Số lượng đã bán"] / df["Số lượng hàng tồn kho"]
    ).replace([np.inf, -np.inf], np.nan)

    # 🔥 MỚI – điểm đánh giá có trọng số
    df["Rating_weighted"] = df["Điểm đánh giá"] * np.log1p(df["Số lượng đánh giá"])

    return df


def aggregations(df):
    avg_disc_by_cat = df.groupby("Danh mục")["Giảm giá"].mean().reset_index().rename(columns={"Giảm giá":"Trung bình giảm giá (%)"}).sort_values("Trung bình giảm giá (%)", ascending=False)
    avg_disc_by_brand = df.groupby("Thương hiệu")["Giảm giá"].mean().reset_index().rename(columns={"Giảm giá":"Trung bình giảm giá (%)"}).sort_values("Trung bình giảm giá (%)", ascending=False)
    avg_disc_by_season = df.groupby("Mùa phổ biến")["Giảm giá"].mean().reset_index().rename(columns={"Giảm giá":"Trung bình giảm giá (%)"}).sort_values("Trung bình giảm giá (%)", ascending=False)

    top10_best_sellers = df.sort_values("Số lượng đã bán", ascending=False).head(10)[["Mã sản phẩm","Tên sản phẩm","Thương hiệu","Danh mục","Số lượng đã bán","Giá bán hiện tại","Tổng doanh thu"]]
    top10_most_loss = df.sort_values("Tổng lãi/lỗ").head(10)[["Mã sản phẩm","Tên sản phẩm","Thương hiệu","Danh mục","Giá gốc","Giá bán hiện tại","Số lượng đã bán","Tổng lãi/lỗ"]]

    disc_75 = df["Giảm giá"].quantile(0.75)
    qty_75 = df["Số lượng đã bán"].quantile(0.75)
    promo_sensitive = df[(df["Giảm giá"] >= disc_75) & (df["Số lượng đã bán"] >= qty_75)].sort_values(["Giảm giá","Số lượng đã bán"], ascending=[False,False])

    brand_agg = df.groupby("Thương hiệu").agg(
        Tổng_doanhthu=("Tổng doanh thu","sum"),
        Trung_binh_giamgia=("Giảm giá","mean"),
        Số_sản_phẩm=("Mã sản phẩm","count")
    ).reset_index().sort_values("Tổng_doanhthu", ascending=False)

    rev_75 = brand_agg["Tổng_doanhthu"].quantile(0.75)
    disc_median = brand_agg["Trung_binh_giamgia"].median()
    strong_brands = brand_agg[(brand_agg["Tổng_doanhthu"] >= rev_75) & (brand_agg["Trung_binh_giamgia"] <= disc_median)]

    cat_agg = df.groupby("Danh mục").agg(
        Tổng_doanhthu=("Tổng doanh thu","sum"),
        Trung_binh_bien_lai=("Biên lợi nhuận","mean"),
        Số_sản_phẩm=("Mã sản phẩm","count")
    ).reset_index()

    rev_median_cat = cat_agg["Tổng_doanhthu"].median()
    margin_median_cat = cat_agg["Trung_binh_bien_lai"].median()
    underperforming_cats = cat_agg[(cat_agg["Tổng_doanhthu"] < rev_median_cat) & (cat_agg["Trung_binh_bien_lai"] < margin_median_cat)].sort_values("Tổng_doanhthu")

    return {
        "avg_disc_by_cat": avg_disc_by_cat,
        "avg_disc_by_brand": avg_disc_by_brand,
        "avg_disc_by_season": avg_disc_by_season,
        "top10_best_sellers": top10_best_sellers,
        "top10_most_loss": top10_most_loss,
        "promo_sensitive": promo_sensitive,
        "brand_agg": brand_agg,
        "strong_brands": strong_brands,
        "cat_agg": cat_agg,
        "underperforming_cats": underperforming_cats
    }


def correlations(df):
    results = {}

    pairs = {
        "discount_vs_sales": ("Giảm giá","Số lượng đã bán"),
        "view_vs_sales": ("Số lượt xem sản phẩm","Số lượng đã bán"),
        "rating_weighted_vs_revenue": ("Rating_weighted","Tổng doanh thu"),
        "conversion_vs_discount": ("Giảm giá","Tỷ lệ chuyển đổi"),
        "stock_pressure_vs_sales": ("Áp lực tồn kho","Số lượng đã bán")
    }

    for name, (x, y) in pairs.items():
        tmp = df[[x,y]].dropna()
        if len(tmp) >= 2:
            r, p = pearsonr(tmp[x], tmp[y])
            rs, ps = spearmanr(tmp[x], tmp[y])
            results[name] = {
                "pearson_r": r, "pearson_p": p,
                "spearman_r": rs, "spearman_p": ps
            }
        else:
            results[name] = None

    return results


def make_plots(df):
    plt.style.use("ggplot")
    plt.figure(figsize=(6,4))
    plt.scatter(df["Giảm giá"], df["Số lượng đã bán"])
    plt.xlabel("Giảm giá (%)")
    plt.ylabel("Số lượng đã bán")
    plt.title("Giảm giá vs Số lượng đã bán")
    plt.show()

    plt.figure(figsize=(6,4))
    plt.scatter(df["Điểm đánh giá"], df["Tổng doanh thu"])
    plt.xlabel("Điểm đánh giá")
    plt.ylabel("Tổng doanh thu")
    plt.title("Điểm đánh giá vs Tổng doanh thu")
    plt.show()

    plt.figure(figsize=(6,4))
    plt.scatter(df["Giá trị định lượng"], df["Giá bán hiện tại"])
    plt.xlabel("Giá trị định lượng")
    plt.ylabel("Giá bán hiện tại")
    plt.title("Giá trị định lượng vs Giá bán")
    plt.show()

    top_by_rev = df.sort_values("Tổng doanh thu", ascending=False).head(10)
    plt.figure(figsize=(8,5))
    plt.barh(top_by_rev["Tên sản phẩm"].str.slice(0,40)[::-1], top_by_rev["Tổng doanh thu"][::-1])
    plt.xlabel("Tổng doanh thu")
    plt.title("Top 10 sản phẩm theo Tổng doanh thu")
    plt.tight_layout()
    plt.show()


def print_summary(aggs, corrs):
    print("\n--- Tóm tắt nhanh ---")
    print("\nGiảm giá trung bình theo danh mục:")
    display(aggs["avg_disc_by_cat"])

    print("\nTop 10 sản phẩm bán chạy nhất (theo Số lượng):")
    display(aggs["top10_best_sellers"])

    print("\nTop 10 sản phẩm lỗ nhiều nhất:")
    display(aggs["top10_most_loss"])

    print("\nSản phẩm nhạy cảm khuyến mãi (>= 75th percentile):")
    if aggs["promo_sensitive"].empty:
        print("Không có sản phẩm thỏa điều kiện.")
    else:
        display(aggs["promo_sensitive"][["Mã sản phẩm","Tên sản phẩm","Giảm giá","Số lượng đã bán","Tổng doanh thu"]])

    print("\n--- Correlation results ---")
    for k, c in corrs.items():
        print(f"{k}: {c}")

print("Đang đọc dữ liệu...")
df = read_and_clean(INPUT_CSV)

print("Tính toán các chỉ số...")
df = compute_metrics(df)

print("Tổng hợp báo cáo...")
aggs = aggregations(df)
corrs = correlations(df)

print("Hiển thị biểu đồ...")
make_plots(df)

print_summary(aggs, corrs)
print("\n✅ Hoàn tất phân tích.")
```

## Phần 4

```python
# PHÂN TÍCH KHUYẾN MÃI & GỢI Ý MARKETING

import numpy as np
import pandas as pd

df_recommend = df.copy()

disc_high = df_recommend["Giảm giá"].quantile(0.75)
sales_high = df_recommend["Số lượng đã bán"].quantile(0.75)
sales_low = df_recommend["Số lượng đã bán"].quantile(0.25)

# --- Hàm suy luận xu hướng ---
def predict_trend(row):
    if row["Giảm giá"] >= disc_high and row["Số lượng đã bán"] >= sales_high:
        return "Tăng trưởng"
    elif row["Giảm giá"] < disc_high and row["Số lượng đã bán"] >= sales_high:
        return "Ổn định"
    elif row["Giảm giá"] >= disc_high and row["Số lượng đã bán"] <= sales_low:
        return "Giảm sút"
    else:
        return "Không rõ"

# --- Hàm đề xuất chiến lược ---
def suggest_strategy(row):
    suggestion = ""
    
    # Dựa vào độ nhạy giá
    if row["Giảm giá"] >= disc_high and row["Số lượng đã bán"] >= sales_high:
        suggestion += "Đề xuất giảm nhẹ giá (5-10%) để tăng thêm doanh số. "
    elif row["Giảm giá"] < disc_high and row["Số lượng đã bán"] >= sales_high:
        suggestion += "Giữ nguyên giá, tập trung quảng bá thương hiệu. "
    elif row["Giảm giá"] >= disc_high and row["Số lượng đã bán"] <= sales_low:
        suggestion += "Giảm giá không hiệu quả, xem xét thu hẹp hoặc thay đổi sản phẩm. "
    
    # Dựa vào mùa phổ biến
    season = str(row.get("Mùa phổ biến", "")).lower()
    if "hè" in season or "thu" in season:
        suggestion += "Đẩy mạnh bán hàng từ tháng 4-9. "
    elif "đông" in season or "tết" in season:
        suggestion += "Đẩy mạnh bán hàng từ tháng 10-2. "
    elif "4 mùa" in season:
        suggestion += "Có thể quảng bá quanh năm, ưu tiên chiến dịch dài hạn. "
        
    if row["Số lượt xem sản phẩm"] > df["Số lượt xem sản phẩm"].quantile(0.75) \
    and row["Tỷ lệ chuyển đổi"] < df["Tỷ lệ chuyển đổi"].median():
        suggestion += "Sản phẩm nhiều lượt xem nhưng tỷ lệ mua thấp → tối ưu giá, hình ảnh hoặc mô tả. "

    if row["Áp lực tồn kho"] > 1:
        suggestion += "Nguy cơ thiếu hàng → cần nhập thêm hoặc giảm quảng cáo. "

    if row["Rating_weighted"] < df["Rating_weighted"].median():
        suggestion += "Ít đánh giá tích cực → khuyến khích khách hàng để lại review. "

    
    # Dựa vào thương hiệu
    brand_sales = df_recommend.groupby("Thương hiệu")["Số lượng đã bán"].mean()
    brand_avg = brand_sales.mean()
    brand_val = brand_sales.get(row["Thương hiệu"], brand_avg)
    
    if brand_val >= brand_avg:
        suggestion += "Thương hiệu mạnh → ra mắt sản phẩm mới hoặc giữ giá. "
    else:
        suggestion += "Thương hiệu yếu → khuyến khích combo với thương hiệu mạnh hoặc quảng cáo thêm. "
    
    # Dựa vào danh mục
    cat_sales = df_recommend.groupby("Danh mục")["Số lượng đã bán"].mean()
    cat_avg = cat_sales.mean()
    if cat_sales.get(row["Danh mục"], cat_avg) < cat_avg:
        suggestion += "Danh mục đang kém → cần chiến dịch quảng cáo hoặc giảm giá kích cầu. "
    
    return suggestion.strip()

df_recommend["Xu hướng"] = df_recommend.apply(predict_trend, axis=1)
df_recommend["Gợi ý chiến lược"] = df_recommend.apply(suggest_strategy, axis=1)

output_file = "marketing_recommendations.csv"
df_recommend.to_csv(output_file, index=False, encoding='utf-8-sig')

print(f"\n✅ Đã xuất file gợi ý chiến lược → {output_file}")
print("Các cột gồm: Mã sản phẩm, Tên sản phẩm, Xu hướng, Gợi ý chiến lược\n")

display(df_recommend[[
    "Mã sản phẩm", "Tên sản phẩm", "Danh mục", "Thương hiệu",
    "Giảm giá", "Số lượng đã bán", "Mùa phổ biến",
    "Xu hướng", "Gợi ý chiến lược"
]].head(10))
```
